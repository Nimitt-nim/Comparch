```python
import matplotlib.pyplot as plt
```

```python
# Plotting
# Datatype = int

N = [64, 128, 256, 512, 1024]
user_times = [0.0086, 0.0412, 0.1588, 0.9671, 8.198]
system_times = [0.0039, 0.0042, 0.0049, 0.0068, 0.0340]

# Plot for User CPU Times
plt.figure(figsize=(8, 6))
plt.plot(N, user_times, marker='o', color='b', label='User CPU Time')
plt.title('User CPU Time vs Matrix Size')
plt.xlabel('Matrix Size (N)')
plt.ylabel('User CPU Time (seconds)')
plt.xticks(N)
plt.grid(True)
plt.legend()
plt.show()

# Plot for System CPU Times
plt.figure(figsize=(8, 6))
plt.plot(N, system_times, marker='o', color='r', label='System CPU Time')
plt.title('System CPU Time vs Matrix Size')
plt.xlabel('Matrix Size (N)')
plt.ylabel('System CPU Time (seconds)')
plt.xticks(N)
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for int
N_int = [64, 128, 256, 512, 1024]
cpu_time_int = [0.0022, 0.00512, 0.0490, 0.3730, 3.134]

# Data for long
N_long = [64, 128, 256, 512, 1024]
cpu_time_long = [0.0021, 0.0059, 0.0490, 0.3830, 3.347]

# Plot for int data type
plt.figure(figsize=(8, 6))
plt.plot(N_int, cpu_time_int, marker='o', color='b', label='Int DataType')
plt.title('CPU Time vs Matrix Size (Int DataType)')
plt.xlabel('Matrix Size (N)')
plt.ylabel('CPU Time (seconds)')
plt.xticks(N_int)
plt.grid(True)
plt.legend()
plt.show()

# Plot for long data type
plt.figure(figsize=(8, 6))
plt.plot(N_long, cpu_time_long, marker='o', color='r', label='Long DataType')
plt.title('CPU Time vs Matrix Size (Long DataType)')
plt.xlabel('Matrix Size (N)')
plt.ylabel('CPU Time (seconds)')
plt.xticks(N_long)
plt.grid(True)
plt.legend()
plt.show()

# Combined plot for comparison
plt.figure(figsize=(8, 6))
plt.plot(N_int, cpu_time_int, marker='o', color='b', label='Int DataType')
plt.plot(N_long, cpu_time_long, marker='o', color='r', label='Long DataType')
plt.title('CPU Time vs Matrix Size (Int vs Long DataType)')
plt.xlabel('Matrix Size (N)')
plt.ylabel('CPU Time (seconds)')
plt.xticks(N_int)
plt.grid(True)
plt.legend()
plt.show()
```
