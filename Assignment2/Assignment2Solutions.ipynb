**Question 1**

Suppose that a task makes extensive use of floating-point operations, with 40% of the time consumed by floating-point operations. With a new hardware design, the floating-point module is sped up by a factor of K. Then what is the overall speedup?

Solution

We know,

$$T_{original} = T_{p} + T_{1-p} \\[2em]

T_{new} = \frac{T_{p}}{N} + T_{1-p} \\[2em]

speedup(new, original) = \frac{T_{original}}{T_{new}} \\[2em]
\frac{T_{p}}{p} = \frac{T_{1-p}}{1-p} = T_{original}$$


Given, $$p = 0.4 \\ N = K$$

$$$$

Plugging values, 

$$speedup = \frac{T_{p} + T_{1-p}}{\frac{T_{p}}{N} + T_{1-p}}\\[2em]

 = \frac{p*T_{original} + (1-p)T_{original}}{\frac{p*T_{original}}{N} + (1-p)T_{original}}\\[2em]

 = \frac{1}{\frac{p}{N} + (1-p)}\\[2em]

 = \frac{1}{\frac{0.4}{N} + (1-0.4)}\\[2em]

 = \mathbf{\frac{N}{0.4 + 0.6*N}}$$

**Question 3**

Consider a system that has a single bottleneck which occupies 20% of the total execution time. Suppose we add 4 more processors to the system. What would be the speedup achieved?


Solution 

Given, $$1-p = 0.2 \\ N = 4$$

Solving, $ p = 0.8$, using the equation derived in Question 1,

$$speedup = \frac{1}{\frac{p}{N} + (1-p)} = \frac{1}{\frac{0.8}{4} + (0.2)} = \mathbf{2.5}$$


**Question 4**

A system is composed of 4 components:

a). The performance of 5% of the system can be doubled. We will call this part component 1

b). The performance of 20% of the system can be improved by 80%. We will call this part component 2

c). The performance of 45% of the system can be improved by 50%. We will call this part component 3

d). The performance of the remaining part of the system cannot be improved. We will call this part 
component 4.

Which component is most worthy to work on to get the maximum overall improvement?

Solution 

We can only improve three components : Component 1, 2 and 3. We compute speedup for each of the cases and report component with highest speedup most worhty to work on.

We know, $speedup = \frac{1}{\frac{p}{N} + (1-p)}$

|Case|Component|p|N|speedup calculation| speedup|

|I|1|0.05|2|$\frac{1}{\frac{0.05}{2} + (0.95)}$|1.02|

**Question 5**

Assume a machine is enhanced, making all floating-point instructions 7 times faster. Originally,
these floating-point instructions accounted for 65% of the total execution time.

a). What will be the overall speedup of the machine after this enhancement?
Additionally, calculate the percentage increase in speedup due to this change.

b). If the execution time of some benchmark program before the floating-point enhancement is 29 seconds, what will the speedup be if only two-third of the 29 seconds is spent executing floating-point instructions.

Solution 


(a) We know, $$speedup = \frac{1}{\frac{p}{N} + (1-p)}$$

Given, $$p = 0.65 \\ N = 7$$

Plugging values, we have,

$$SpeedUp = \frac{1}{\frac{0.65}{7} + (0.35)} = \mathbf{2.26}$$

Assuming percentage increase in speedup refers to percentage increase in performance, 

$$
Performance_{new} = 2.26 \times Performance_{original}\\[2em]

Performance_{new} - Performance_{original} = 1.26 \times Performance_{original}\\[2em]

\frac{(Performance_{new} - Performance_{original})}{Performance_{original}} \times 100 = 1.26 \times 100\\[2em]

Percentage \ Increase = \mathbf{126} \%
$$


(b) Given,
$$
T_{original} = 29 \ s \\[2em]

T_{new} = [\frac{1}{7}\times\frac{2}{3}\times29 + \frac{1}{3}\times29 = \frac{9}{21}\times29] \ s \\[2em]

speedup = \frac{T_{original}}{T_{new}} = \frac{29}{\frac{9}{21}\times29} = \frac{21}{9} = \mathbf{2.33} \\

**Question 6**

 When parallelizing an application, the ideal speedup is speeding up by the number of processors. This is limited by two things: percentage of the application that can be parallelized and the cost of communication.Amdahlâ€™s law takes into account the former but not the latter.
 
a. What is the speedup with N processors if 80% of the application is parallelizable, ignoring the cost of communication?

b. What is the speedup with 8 processors if, for every processor added, the communication overhead is 0.5% of the original execution time.

c. What is the speedup with 8 processors if, for every time the number of processors is doubled, the communication overhead increases by 0.5% of the original execution time?

d. What is the speedup with N processors if, for every time the number of processors is doubled, the communication overhead increases by 0.5% of the original execution time?

e. Write the general equation that solves this question: What is the number of processors with the highest speedup in an application in which P% of the original execution time is parallelizable, and, for every time the number of processors is doubled, the communication is increased by 0.5% of the original execution time?

Solution

(a) We know, $$speedup = \frac{1}{\frac{p}{N} + (1-p)}$$

Given, $$p = 0.8 \\ N = N$$

Plugging values, we have,

$$SpeedUp = \frac{1}{\frac{0.8}{N} + (0.2)} = \mathbf{\frac{N}{0.8 + 0.2\times N}}$$

(b) We know,

$$T_{original} = T_{p} + T_{1-p} \\[2em]

T_{new} = \frac{T_{p}}{N} + T_{1-p} + N\times 0.005 \times T_{original}\\[2em]

speedup(new, original) = \frac{T_{original}}{T_{new}} \\[2em]
\frac{T_{p}}{p} = \frac{T_{1-p}}{1-p} = T_{original}$$


Given, $$p = 0.8 \ (assuming \ given \ same \ as \ (a)) \\ N = 8$$

$$$$

Plugging values, 

$$speedup = \frac{T_{p} + T_{1-p}}{\frac{T_{p}}{N} + T_{1-p}+ N\times 0.005\times T_{original}} \\[2em]

 = \frac{p*T_{original} + (1-p)T_{original}}{\frac{p*T_{original}}{N} + (1-p)T_{original} + N\times 0.005\times T_{original}}\\[2em]

 = \frac{1}{\frac{p}{N} + (1-p) + N\times 0.005}\\[2em]

 = \frac{1}{\frac{0.8}{8} + 0.2 + 8\times 0.005}\\[2em]

 = {\frac{1}{0.34}}\\[2em]

 = \mathbf{2.94}$$


(c) Given, $$p = 0.8 \ (assuming \ given \ same \ as \ (a)) \\ N = 8$$

We can note, N = 8 implies that number of processors is doubled 3 times. The question statement is ambiguous, we can assume that the communication overhead increases by 0.5 % of the oldest original execution time instead immediate orginal.

$$T_{original} = T_{p} + T_{1-p} \\[2em]

T_{new} = \frac{T_{p}}{8} + T_{1-p} + 3\times 0.005 \times T_{original}$$

$$speedup = \frac{T_{p} + T_{1-p}}{\frac{T_{p}}{8} + T_{1-p}+ 3\times 0.005\times T_{original}} \\[2em]

 = \frac{p*T_{original} + (1-p)T_{original}}{\frac{p*T_{original}}{8} + (1-p)T_{original} + 3\times 0.005\times T_{original}}\\[2em]

 = \frac{1}{\frac{p}{8} + (1-p) + 3\times 0.005}\\[2em]

 = \frac{1}{\frac{0.8}{8} + 0.2 + 3\times 0.005}\\[2em]

 = {\frac{1}{0.315}}\\[2em]

 = \mathbf{3.17}$$


(d) Given, $$p = 0.8 \ (assuming \ given \ same \ as \ (a)) \\ N = N$$

We can note, N = N implies that number of processors is doubled $\lfloor \log_2(N) \rfloor$ times. The question statement is ambiguous, we can assume that the communication overhead increases by 0.5 % of the oldest original execution time instead immediate orginal. Also, the overhead only increases upon doubling not addition of non doubling processors.

$$T_{original} = T_{p} + T_{1-p} \\[2em]

T_{new} = \frac{T_{p}}{N} + T_{1-p} + \lfloor \log_2(N) \rfloor\times 0.005 \times T_{original}$$

$$speedup = \frac{T_{p} + T_{1-p}}{\frac{T_{p}}{N} + T_{1-p}+ \lfloor \log_2(N) \rfloor\times 0.005\times T_{original}} \\[2em]

 = \frac{p*T_{original} + (1-p)T_{original}}{\frac{p*T_{original}}{N} + (1-p)T_{original} + \lfloor \log_2(N) \rfloor\times 0.005\times T_{original}}\\[2em]

 = \frac{1}{\frac{p}{N} + (1-p) + \lfloor \log_2(N) \rfloor\times 0.005}\\[2em]

 = \mathbf{\frac{1}{\frac{0.8}{N} + 0.2 + \lfloor \log_2(N) \rfloor\times 0.005}}$$


 (e) As derived in (d), 

 $$ speedup(p,N) = \frac{1}{\frac{p}{N} + (1-p) + \lfloor \log_2(N) \rfloor\times 0.005}$$
